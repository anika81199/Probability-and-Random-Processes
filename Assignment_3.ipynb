**Q.2.** (a) Y=exp(X)

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math

#5 different values of lambda
mean=[0.01,0.05,0.10,0.15,0.20]

for i in mean:
  #generating exponential distribution for different mean
  X=np.random.exponential(i,10000)
  fig, ax = plt.subplots(figsize=(6, 3))

  #defining RV Y~exp(X)
  Y=[]
  for j in X:
    y=np.exp(j)
    Y.append(y)
  plt.hist(Y,bins=100,label='Simulated')
  plt.title("PDF of Y with mean= "+str(i))
  plt.xlabel("y")
  plt.ylabel("fY(y)")
  plt.show()
  print("")

```

**Q.2.** (b) Z=min(X,3)

```python
#for Z ~ min(X,3)
mean=[1,1.2,1.5,2,2.20]
for i in mean:
  #generating exponential distribution for different mean
  X=np.random.exponential(i,10000)
  fig, ax = plt.subplots(figsize=(6, 3))
  Z=[]
  #defining RV Z~min(X,3)
  for j in X:
    if(j<3):
      Z.append(j)
    else:
      Z.append(3)
  plt.hist(Z,bins=100,density=True)
  plt.title("PDF of Z with mean= "+str(i))
  plt.xlabel("z")
  plt.ylabel("fZ(z)")
  plt.show()
  print(" ")
```

**Q.3.** (a) Y=X^(3)

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math

#generating X=Unif(-2pi,2pi)
X=np.random.uniform(-2*math.pi,2*math.pi,10000000)
plt.hist(X,bins=100,density=True)
plt.title("PDF of X")
plt.xlabel("x")
plt.ylabel("fX(x)")
plt.show()

#generating Y~X^3
Y=[]
for i in X:
  y=math.pow(i,3)
  Y.append(y)
plt.hist(Y,bins=100,density=True,label="Simulated")
plt.xlabel("y")
plt.ylabel("fY(y)")
plt.title("PDF of Y")

z=np.arange(1,max(Y),0.1)
y=1/(12*math.pi*pow(z,(2/3)))
plt.plot(-z,y,label='Analytical',color='r')
plt.plot(z,y,color='r')
plt.legend();
plt.show()
```

**Q.5.** (a) X  Exp()

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
#generating exponential distribution
X=np.random.exponential(1,10000)
plt.hist(X,bins=100,density=True)
plt.title("PDF of X")
plt.xlabel("x")
plt.ylabel("fX(x)")
plt.show()

Y=[]
#Y~FX(X)
for x in X:
  y=1-math.exp(-x)
  Y.append(y)

plt.hist(Y,bins=1000,density=True,label="CDF of X or PDF of Y")
plt.legend();

plt.title("Distribution of Y")
plt.xlabel("y")
plt.ylabel("fY(y)")



CDF=[]
for x in X:
  y=1-math.exp(-x)
  z=math.log(1/(1-y))
  res=1-math.exp(math.log(1-y))
  CDF.append(res)
plt.hist(CDF,bins=100,density=True,label="CDF of Y",color='r')
plt.legend();
plt.show()

#they are coming out to be same
```

**Q.6.** Random 200 points in the interval (0, 100)

```python
N = 1000
Z = np.zeros(N)
for i in range(N):
  n = 200
  X = np.random.uniform(0,100,n)
  Z[i] = min(X)

plt.figure(figsize=(4,4))
plt.subplot(1,1,1)
plt.hist(Z,bins=100,density=True)
plt.tight_layout()
plt.show()
```

**Q.7.**

```python
import random
choice=[0,1]

user= int(input())
X=[]
Y=[]
for j in range(1,user+1):
  i=0
  store=[]
  while(i!=1):
    i=random.choice(choice)
    if i==1:
      store.append('H')
    else:
      store.append('T')
  print("Round "+str(j),store)
  print("toses in this round=",len(store))
  print("")
  X.append(len(store))
print("X= ",X)

for i in X:
  y=abs(i-5)
  Y.append(y)
print("Y=",Y)
pmf=[]
temp=[]
for element in Y:
    if element not in temp:
      temp.append(element)

for i in temp:
    freq=0
    for j in Y:
      if(i==j):
        freq=freq+1
    pmf.append(freq/len(Y))
print("PMF of Y=",pmf)
```

we can check the pmf of Y is correct by adding all its entries, which is resulting in 1.
